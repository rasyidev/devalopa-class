```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.figure(figsize=(12,8))
plt.plot([1.6, 2.7])
```

```python
# Membuat angka 0 - 12 dibagi menjadi 70 bagian sama besar
x = np.linspace(0, 12, 70)
print(x)
```

```python
plt.plot(x**3)
```

```python
plt.plot(np.sin(x))
```

```python
plt.plot(np.cos(x))
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.sin(x), color='red')
plt.plot(np.cos(x), 'g--')
plt.plot(np.sqrt(x))
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.sin(x), color='red', label="SIN")
plt.plot(np.cos(x), 'g--', label="COS")
plt.plot(np.sqrt(x), label="Akar")
plt.legend(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.sin(x), color='red', label="SIN")
plt.plot(x**3, color='orange', label= "Qubic")
plt.plot(np.cos(x), 'g--', label="COS")
plt.plot(np.sqrt(x), label="Akar")
plt.xlim(1,70)
plt.ylim(-1, 4)
plt.legend(fontsize=15, loc='upper right')
plt.show()
```

## Subplot: Membuat beberapa visualisasi data dalam satu gambar

```python
# Membuat subplot yang berisi 1 baris dan 2 kolom
# figure: gambar
# axis: koordinat

figure, axis = plt.subplots(nrows=1, ncols=2, figsize=(13,5))
axis[0].plot(np.sin(x), label="SIN")
axis[0].plot(x**2, label="Power")
axis[0].set_xlim(0,70)
axis[0].set_ylim(-1,4)
axis[0].set_title("Grafik A")
axis[0].legend(loc="upper right")

axis[1].plot(np.cos(x), label="COS")
axis[1].set_title("Grafik B")
axis[1].legend(loc="upper right")
plt.show()
```
