### .loc dan .iloc
 **FORMAT** 
 
 ```python
     df.loc[baris, kolom] # index berupa string
     df.iloc[baris, kolom] # index berupa int (dari 0)
 ```
 ```
 awal:akhir
 ```


````python
df.iloc[1, [1,3,5]]
````

````python
df.iloc[0, 1]
````

````python
df.loc[1:2, 'so2':'max']
````

````python
df.loc[4, 'critical':'location']
````

````python
df.iloc[4, 7:]
````

````python
df_baru = pd.DataFrame({
    "a": [11,12,13],
    "b": [14,15,16],
    "c": [17,18,19]
})
````

````python
# Gabung data vertikal
pd.concat([df_buatan_sendiri, df_baru], axis=0)
````

````python
# Gabung data horizontal
pd.concat([df_buatan_sendiri, df_baru], axis=1)
````
